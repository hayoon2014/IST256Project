```python
try:
    response = requests.get("http://www.syrairport.org/flightdata/data.xml")
    with open ("flightdata.xml", "w") as file:
        file.write(response.text)
        
except requests.exceptions.HTTPError as e:
    print("ERROR: Syracuse Airport HTTP is unresponsive.")
    print("DETAILS:", e)

except requests.exceptions.RequestException as e: 
    print("ERROR: Cannot connect to internet")
    print("DETAILS:", e)
```

```python
with open("flightdata.xml") as file:
    doc = xmltodict.parse(file.read())
```

```python
all_flights = doc['flights']['flight']
```

```python
all_flights
```

```python
flight_number = input("Enter your flight number for details! ").upper()
count = 0

for flight in all_flights:
    
    if (flight['@airlinecode']+flight['@flightnumber']) == flight_number:
        count = count + 1

        if flight['@remarks'] == 'CANCELED':
            print("Your flight %s has been canceled. " %flight_number)
            
        elif flight['@type'] == 'A':
            print("Your flight %s is arriving from %s." %(flight_number, flight['@city']))
            print("It is scheduled to arrive %s; projected actual time is %s." %(flight['@scheduletime'], flight['@actualtime']))
            
        elif flight['@type'] == 'D':
            print("Your flight %s is headed to %s." %(flight_number, flight['@city']))
            print("It is scheduled to depart %s; projected actual time is %s." %(flight['@scheduletime'], flight['@actualtime']))

if count == 0:
    print("I'm sorry, that flight isn't arriving to or departing from Hancock International Airport! ")
```
